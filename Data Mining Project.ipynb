```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = pd.read_csv('tips.csv')
tips.tail()
```

```python
tips.info()
```

```python
sns.boxplot('total_bill', data=tips)
```

```python
plt.figure(figsize=(13,8))
corr = tips.corr()
sns.heatmap(corr[(corr > 0.2) | (corr <= -1)], 
            cmap='Reds', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 14}, square=True)
plt.title('Korelasi High_Price',fontsize=30)
plt.tight_layout();
```

Berdasarkan heatmap diatas dapat dilihat bahwa korelasi antara berapa banyak orang per meja mempengaruhi harga total dari pembayaran

```python
sns.catplot(x='sex',y='total_bill',data=tips)
plt.title('Perbandingan Ukuran Gender Dalam Pembayaran',fontsize=15)
plt.tight_layout()
plt.xlabel('Sex')
plt.ylabel('Total Bill')
```

Berdasarkan penyebaran data disini dapat dilihat bahwa pada gender laki laki sering kali mendapatkan cost total yang sangat besar

```python
p = sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue = tips['sex'])
```

Berdasarkan scatterplot diatas bahwa kebanyakan customer memberikan tip dengan total bill rentang 10-20 dollar dan memberikan tip mulai dari 1-4 dollar.

```python
q = sns.scatterplot(x=tips['total_bill'], y=tips['day'], hue = tips['sex'])
```

Berdasarkan grafik diatas bahwa orang banyak membeli pada hari weekend
