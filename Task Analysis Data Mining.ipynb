```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Mengintip beberapa data pertama
rumah_bekasi = pd.read_excel('Dataset Tugas 1.xlsx')
rumah_bekasi.head()
```

# Exploratory Data Analysis

```python
# Mengecek Info Data
rumah_bekasi.info()
```

```python
# Cek data duplikat
rumah_bekasi.duplicated().any()
```

###### * Dari hasil pengecekan tidak terdapat data duplikat

```python
# Ubah tipe data tanggal dari object ke date time
rumah_bekasi['Tgl_Iklan']=pd.to_datetime(rumah_bekasi['Tgl_Iklan'],dayfirst=True)
# Mengurutkan data sesuai tanggal
rumah_bekasi.sort_values('Tgl_Iklan', ascending=True, inplace= True)
rumah_bekasi.head()
```

```python
#Cek Tipe Data yang diubah
rumah_bekasi['Tgl_Iklan'].dtypes
```

```python
# Drop kolom yang tidak memiliki hubungan
rumah_bekasi.drop('No',axis=1, inplace=True)
rumah_bekasi.drop('Tgl_Iklan',axis=1, inplace=True)
rumah_bekasi.drop('url',axis=1, inplace=True)
rumah_bekasi.drop('judul',axis=1, inplace=True)
rumah_bekasi.drop('deskripsi',axis=1, inplace=True)
```

```python
# Cek data kembali
pd.options.display.max_rows=100
pd.options.display.max_columns=100
rumah_bekasi
```

```python
# Cek missing value
rumah_bekasi.isnull().sum()
```
